## String (in Python)

### 1Ô∏è‚É£ What is a String?

A **string** is a collection of characters (alphabets, digits, symbols, spaces) used to form a word, sentence, or paragraph.

In Python, strings are enclosed in:

* **Single quotes** ‚Üí `'hello'`
* **Double quotes** ‚Üí `"hello"`
* **Triple single quotes** ‚Üí `'''long string'''`
* **Triple double quotes** ‚Üí `"""long string"""`

Triple quotes are used for **multi-line strings**.

---

### 2Ô∏è‚É£ String Representation (Memory & Indexing)

Example:

```python
s = 'hello'
```

Each character is stored at a specific index (position):

| Character | h | e | l | l | o |
| --------- | - | - | - | - | - |
| Index     | 0 | 1 | 2 | 3 | 4 |

So,

```python
print(s[1])   # Output: e
print(s[4])   # Output: o
```

---

### 3Ô∏è‚É£ String Indexing (`[]`)

* Used to retrieve a single character.
* Index starts from **0**.
* Negative indexing is also allowed:

| Character | h  | e  | l  | l  | o  |
| --------- | -- | -- | -- | -- | -- |
| Positive  | 0  | 1  | 2  | 3  | 4  |
| Negative  | -5 | -4 | -3 | -2 | -1 |

Example:

```python
print(s[-1])   # Output: o
```

---

### 4Ô∏è‚É£ String Slicing (`[start:stop:step]`)

Used to retrieve multiple characters.

Syntax:

```python
s[start:stop:step]
```

* **start** ‚Üí starting index (included)
* **stop** ‚Üí ending index (excluded)
* **step** ‚Üí jump value

Examples:

```python
print(s[1:4])     # ell
print(s[0:5:2])   # hlo
print(s[::-1])    # olleh (reverse string)
```

---

### 5Ô∏è‚É£ String is Immutable (Important)

Once a string is created:

* ‚ùå Characters cannot be changed
* ‚ùå Length cannot be modified

Example:

```python
s = "hello"
s[0] = 'H'   # ‚ùå Error (Strings are immutable)
```

So, **String is immutable** in Python.

---

### 6Ô∏è‚É£ Operations on Strings

### üîπ 1. Concatenation (`+`)

Joining two strings.

```python
a = "hello"
b = "world"
print(a + b)   # helloworld
```

---

### üîπ 2. Repetition (`*`)

Repeats string multiple times.

```python
print("hi" * 3)   # hihihi
```

---

### üîπ 3. Membership (`in`, `not in`)

```python
print("h" in s)       # True
print("z" not in s)   # True
```

---

### üîπ 4. String Comparison

Comparison operators:

* `<`
* `<=`
* `>`
* `>=`
* `==`
* `!=`

Strings are compared **character by character (ASCII/Unicode value)**.

```python
print("abc" == "abc")   # True
print("abc" < "abd")    # True
```

---

## ‚úÖ Summary

* String = collection of characters
* Written using `' '`, `" "`, `''' '''`, `""" """`
* Indexing starts from 0
* Slicing ‚Üí `[start:stop:step]`
* Strings are **immutable**
* Supports operations: `+`, `*`, `in`, comparisons
