```python
# Write a program to count how many vowels are in a given sentence (use a loop and .lower()).

# a, e, i, o, u and sometime y.

# Y acts as a vowel in words like sky, myth, or boy, producing an /i/ or /aɪ/ sound.

# Short Vowels: Typically produce a distinct, quick sound (e.g., a in apple, e in bed, i in sit, o in hot, u in cup).
# Long Vowels: The vowel sounds like its letter name (e.g., a in cake, e in feet, i in ice, o in rope, u in cute).

word = input('Enter any word: ')
count = 0 
vowels = ('a', 'e', 'i', 'o', 'u')

for char in word.lower():
    if char in vowels:
        count += 1

print(count)




```

```python
# Write a function classify_score(score) that returns "Fail", "Pass", "Merit", or "Distinction" based on score ranges.

# <40 Fail, 40-59 Pass, 60-79 Merit, 80+ Distinction




def classify_score(score):
    if score < 40:
        return 'Fail'
    elif 40 <= score <= 59:
        return 'Pass'
    elif 60 <= score <= 79:
        return 'Merit'
    else:
        return 'Distinction'

result = classify_score(9)

print(result)


```

```python
# URL Parsing
# URL    --> https://www.kaggle.com/datasets
# Answer --> Protocol: https | Domain: kaggle | Page: datasets

from urllib.parse import urlparse

URL        = 'https://www.kaggle.com/datasets'
parsed_URL = urlparse(URL)

print('Full Parsed URL:', parsed_URL)
print()

# Direct attributes
print(f'Scheme (full) : {parsed_URL.scheme}')   # https
print(f'Netloc (full) : {parsed_URL.netloc}')   # www.kaggle.com
print(f'Path   (full) : {parsed_URL.path}')     # /datasets
print()

# Extract Protocol
Protocol      = parsed_URL.scheme                          # https

# Extract Domain — remove 'www.' from left, '.com' from right
_, _, temp    = parsed_URL.netloc.partition('.')           # kaggle.com
Domain, _, _  = temp.partition('.')                        # kaggle

# Extract Page — remove leading '/'
Page    = parsed_URL.path         # datasets

print(f'Protocol : {Protocol}')
print(f'Domain   : {Domain}')
print(f'Page     : {Page}')
```

```python
# Second mathod

url = 'https://www.kaggle.com/datasets'
# TODO: Find the position of the colon (:) to extract protocol

prot = url.find(':') #--> this will give index of :
protocol=url[:prot]


# TODO: Find the positions of the dots (.) to locate the start/end of the domain

dot1 = int(url.find('.'))
dot2 = int(url.find('.',dot1+1))

domain = url[dot1+1:dot2]


# TODO: Find the slash after the domain and slice to get the page part
slash_find = url.find('/',dot2)
page = url[slash_find:]

print("Protocol:", protocol)
print("Domain:", domain)
print("Page:", page)
```

```python
# Anagrams

# Anagrams means like decimal --> medical --> if I rearrange the alphabets, I can create one more word by the using same alphabets.

string = 'deCiMal'
string_1 = 'mEdiCal'

string = string.casefold() #--> convert into lower case
string_1 = string_1.casefold()#--> convert into lower case

print(string_1, string) #check if both above function are working fine or not

for x in string:
    if x.isalpha():
        if string.count(x)!=string_1.count(x):
            print('Not anagram')
            break
else: 
    print('Anagram')        
```

```python
# Password Reset Program

CORRECT_USERNAME = 'Parveen'
CORRECT_PASSWORD = 'Correct'

User_Name = input('Enter your User Name: ')
Password  = input(f'Enter your Password for {User_Name}: ')

# Check username
if User_Name != CORRECT_USERNAME:
    print('User Name not found.')

# Correct login
elif User_Name == CORRECT_USERNAME and Password == CORRECT_PASSWORD:
    print(f'Login Successful! Welcome {User_Name}.')

# Wrong password
elif User_Name == CORRECT_USERNAME and Password != CORRECT_PASSWORD:
    print('Password is wrong.')

    while True:
        User_input = input('Type "Try Again" to retry or "Reset" to reset password: ')

        # Try Again
        if User_input == 'Try Again':
            Password = input(f'Enter your Password for {User_Name}: ')
            if Password == CORRECT_PASSWORD:
                print(f'Login Successful! Welcome {User_Name}.')
                break
            else:
                print('Wrong Password again.')

        # Reset Password
        elif User_input == 'Reset':
            New_Password   = input('Enter your New Password: ')
            New_Password_1 = input('Confirm your New Password: ')

            if New_Password == New_Password_1:
                CORRECT_PASSWORD = New_Password   # ✅ assign with '='
                print('Password reset Successful! Please login again.')
                break
            else:
                print('Passwords do not match. Try again.')

        else:
            print('Invalid option. Type "Try Again" or "Reset".')


    




```
