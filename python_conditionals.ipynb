```python
# ODD or EVEN

A = int(input("Enter any number for A: "))
B = int(input("Enter any number for B: "))

# Check A
if A % 2 == 0:
    print("Your number for A is an even number")
else:
    print("Your number for A is an odd number")

# Check B
if B % 2 == 0:
    print("Your number for B is an even number")
else:
    print("Your number for B is an odd number")

#Same logic with function

def Check_even_or_odd(num, name):
    if num % 2 ==0:
        print(f"{name} is even number")
    else :
        print(f"{name} is odd number")

Check_even_or_odd(A, "A")
Check_even_or_odd(B, "B")           
```

```python
# Eligible to vote (simple check)

age = int(input("Enter your age as per Aadhaar Card: "))

if age >= 18:
    print("Yes, you can vote")
else:
    print("You are not eligible to vote")


#Same logic with function

def eligible_to_vote(age):
    if age >= 18:
        return "You can vote"
    else:
        return "You are not eligible to vote"

dl_age = int(input("Enter your age as per DL: "))

result = eligible_to_vote(dl_age)
print(result)
```

```python
# Age check for work
# if he/she is below than 18 or above than 60 so he/she not eligible for work

age = int(input("Your age as per adhar card: "))

if age >= 18 and age <= 60:
    print("Eligible")
else:
    print("Not Eligible")


#Same logic with function

def eligible_to_work(age_1):
    if age_1 >= 18 and age_1 <= 60:
        return "You Can Work."
    else:
        return "You are not eligible to Work."
    
Age_1 = int(input("Enter your age as per the adhar card: "))

result = eligible_to_work(Age_1)
print(result)





```

```python
#Vowels or Consonant
#Vowel = {a,e,i,o,u}
#Rest of letter are called as Consonant.

Alphabets = input("Enter any alphabets")

if Alphabets in ("a","e","i","o","u") or Alphabets in ("A","E","I","O","U"):
    print("Vowels")
else :
    print("Consonant") 

 #same with function

def Check_type_of_Alphabest(Alphabet):
    if Alphabet in ("a","e","i","o","u") or Alphabet in ("A","E","I","O","U"):
        return "Vowels"
    else :
        return "Consonant"
    
Your_Input = input("Enter any alphabet: ")

result = Check_type_of_Alphabest(Your_Input)
print(result)


```

```python
# Exam result
# Subjects: Math, Phy, Che
# Max Marks = 100
# Passing Marks = 45

Math = int(input("Enter marks for Math: "))
Phy = int(input("Enter marks for Phy: "))
Che = int(input("Enter marks for Che: "))

# Validate marks range
if not (0 <= Math <= 100 and 0 <= Phy <= 100 and 0 <= Che <= 100):
    print("Invalid marks entered")

# Pass in all subjects
elif Math >= 45 and Phy >= 45 and Che >= 45:
    print("Pass in all subjects")

# Fail in all subjects
elif Math < 45 and Phy < 45 and Che < 45:
    print("Failed in all subjects")

# Partial pass
else:
    print("Failed in one or more subjects")

    
```

```python
# Generate Discount Bill
# When amount of bill is less then 1000 then discount is 10%
# when amount of bill is 1000 or more  but less than 5000 then discount is 15%
# from 5000 to 10000 discount is 20%
# above 10000 discount is 25%

Bill_Amount = float(input())

if Bill_Amount < 1000:
    Discount = Bill_Amount * 10/100
    print(Bill_Amount - Discount)
elif Bill_Amount >= 1000 and Bill_Amount < 5000:
    Discount = Bill_Amount * 15/100
    print(Bill_Amount - Discount)  
elif Bill_Amount >= 5000 and Bill_Amount <= 10000:
      Discount = Bill_Amount * 20/100
      print(Bill_Amount - Discount)
else:
     Discount = Bill_Amount * 25/100
     print(Bill_Amount - Discount)      

# Generate Discount Bill

bill_amount = float(input("Enter bill amount: "))

if bill_amount < 1000:
    discount_rate = 0.10
elif bill_amount < 5000:
    discount_rate = 0.15
elif bill_amount <= 10000:
    discount_rate = 0.20
else:
    discount_rate = 0.25

discount = bill_amount * discount_rate
final_amount = bill_amount - discount

print("Bill Amount:", bill_amount)
print("Discount:", discount)
print("Final Amount to Pay:", final_amount)
```

```python
# Leap Year or Not (Gregorian Calendar)

year = int(input("Enter year: "))

if year % 400 == 0:
    print("Leap Year")
elif year % 100 == 0:
    print("Not a Leap Year")
elif year % 4 == 0:
    print("Leap Year")
else:
    print("Not a Leap Year")
  
```
