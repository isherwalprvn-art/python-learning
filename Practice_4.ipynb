```python
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

work_hours = [int(input(f'Enter working hours for {day}: ')) for day in days]

total  = sum(work_hours)
normal_pay  = 40 * 12
extra_hours = total - 40

if total <= 40:
    pay = total * 12
else:
    pay = normal_pay + extra_hours * 12 * 1.5

print(f'Total hours worked this week: {total}')
print(f'Pay for this week:          £{pay:.2f}')
```

```python
L1 = [3, 5, 7, 9, 3, 6, 5, 2, 3, 7]

# Method 1 — set() shortest but may change order
L2 = sorted(set(L1))

# Method 2 — loop, preserves original order
L2 = []
for l in L1:
    if l not in L2:
        L2.append(l)

# Method 3 — list comprehension
L2 = []
[L2.append(l) for l in L1 if l not in L2]

print(L2)    # → [2, 3, 5, 6, 7, 9]
```

```python
# Rotate a list by 2 places and place them from starting to end 

L1 = [1,2,3,4,5,6]

# Way A
L2 = L1[2:len(L1)] + L1[0:2]

# Way B
L2 = L1[2:] + L1[:2]
```

```python
# Check if a list is palindrom or not

L1 = [5,4,3,3,4,5]

L2 = L1[::-1].copy()

if L1 == L2:
    print('palindrom')
else:
    print('not palindrom')    
```
