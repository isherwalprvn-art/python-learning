# We have to print a restaurant menu in which we have add four items with their corresponding price 
#      Menu
# Hot Dog________$5
# Apple shake____$3
# Panner Burji__$10
# Biryani_______$20

# so length need to be same for all four items we have file the gap between the item name and price with dashes 

```python
# Restaurant Menu

Menu = {
    'Hot Dog' : '$5',
    'Apple Shake' : '$3',
    'Panner Burji' : '$10',
    'Biryani' : '$20'

}

#print(Menu)

print('=' * 25)
print('         Menu')
print('=' * 25)

for item, price in Menu.items():
    print(f'{item:-<20}${price:>3}')
    
print('')
print('='* 25)    

```

```python
# Dynamically find the longest item and set width automatically:

# All our are made in 100% Desi Ghee

Menu = {
    'Chikku Shake' : '$2',
    'Paneer Do Pyaza' : '$8',
    'Paneer Tikka Masala Curry' : '$15',
    'Aalo Puri' : '$10',
    'Rajma Rice Bowl' : '$10'

}

max_lenth = len(max(Menu.keys(), key=len))

#print(max_lenth)

print('=' * (max_lenth+8))
print('Menu'.center((max_lenth+8)))
print('All items made in 100% Desi Ghee'.center(max_lenth + 8))
print('=' * (max_lenth+8))

print('')

for item, price in Menu.items():
    print(f'{item:-<{max_lenth+3}}{price:>3}')

print('')
print('=' * (max_lenth+8))    
```
