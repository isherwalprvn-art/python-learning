```python
# You are given a raw log line from a data pipeline. Extract and print useful information from it.
# Tasks:
# 1. Split the log by " | " and store in a list
# 2. Print each part on a new line with its label
# 3. Check if the word "ERROR" is in the log → print "ALERT: This is an error log!"
# 4. Check if "FAILED" is in the log → print "ALERT: Pipeline failed!"

# Log Part 1: ERROR
# Log Part 2: 2024-01-15
# Log Part 3: pipeline_sales
# Log Part 4: Records processed: 0
# Log Part 5: Status: FAILED
# ---------------------------
# ALERT: This is an error log!
# ALERT: Pipeline failed!

log = "ERROR | 2024-01-15 | pipeline_sales | Records processed: 0 | Status: FAILED"

# log_part_1 = log.split('|')
# print(log_part_1) # To check how spilt is working

log_part = log.split('|') # it wil change it into list

index = 0

while index < len(log_part):
    if  'ERROR' in log_part[index]:
        print('ALERT: This is an error log')
        print(f'Log Part {index+1}: {log_part[index].lstrip()}\n')
    elif 'FAILED' in log_part[index]:
        print('ALERT: Pipeline failed!')
        print(f'Log Part {index+1}: {log_part[index].lstrip() }\n') 
    else :
        print(f'Log Part {index+1}: {log_part[index].lstrip() }\n') 
    index += 1         




```

```python
# Shallow Copy
#  Start with an original list
L1 = [2, 3, 5, 1]

# L2 = L1 is NOT a copy — it's an alias.
# L1 and L2 now point to the exact same list in memory.
L2 = L1

# L3 = L1.copy() creates a brand new list
# with the same values, but stored separately in memory.
L3 = L1.copy()


# Changing L1's first element also changes L2
# (they share the same object), but NOT L3 (it's independent).
L1[0] = 0

# Output:
# L1 → [0, 3, 5, 1]   ← changed
# L2 → [0, 3, 5, 1]   ← also changed (same object as L1)
# L3 → [2, 3, 5, 1]   ← unchanged (its own copy)
print(L1,"\n",L2,"\n",L3)
```

```python
import copy  # need this module for deepcopy

original = [1, [10, 20]]  # a list containing another list inside

# Shallow copy — outer list is new, but inner list is still shared
shallow = original.copy()

# Deep copy — everything is brand new, nothing is shared
deep = copy.deepcopy(original)

# Now change the inner list through original
original[1][0] = 99

print(original)  # [1, [99, 20]]  ← changed
print(shallow)   # [1, [99, 20]]  ← also changed! (shares inner list)
print(deep)      # [1, [10, 20]]  ← untouched (fully independent)
```

```python
cars = ['Tata','BMW','AUDI','Tesla']
# list of 4 cars, index: 0=Tata, 1=BMW, 2=AUDI, 3=Tesla

for i in range(len(cars)):
# range(len(cars)) = range(4) = [0, 1, 2, 3]
# PROBLEM: range is fixed at start — it doesn't shrink
# even if we remove items from the list

    if cars[i].startswith('T'):
    # checks if car name begins with letter 'T'
    # 'Tata'  → True
    # 'BMW'   → False
    # 'AUDI'  → False
    # 'Tesla' → True

        cars.pop(i)
        # removes car at index i
        # BUT list shrinks after each pop!
        # so indexes shift — causes skip or crash

print(cars)
```

```python
cars = ['Tata','BMW','AUDI','Tesla']

# loop over a copy so original can be safely modified
for c in cars.copy():
    if c.startswith('T'):
        cars.remove(c)

print(cars)
# Output: ['BMW', 'AUDI']
```
