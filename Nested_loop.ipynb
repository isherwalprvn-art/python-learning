#Nested Loop
:- Nothing but loop inside loop.
:- Nested are primaryly useful for 2 dimensional data.


```python
for i in range(1,4):
    for j in range(1,4):
        print(i,'-',j,end=' ') #end is like to print the data in tableaur form
```

```python
#Check it is prime number or not

n = 7
count= 0 

for i in range (1,n+1):
    if n % i == 0:
        count += 1
print(i,'-',count)

if count == 2:
    print('Prime Number')
else:
    print('Not a Prime Number')    
```

```python
# Print prime numbers between 1 and 100

for num in range(1, 101):
    count = 0
    
    for i in range(1, num + 1):
        if num % i == 0:
            count += 1
    
    if count == 2:
        print(num)   
      
```
