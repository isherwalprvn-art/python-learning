```python
# addition function
def add(firstnumber, secondnumber):
    return firstnumber + secondnumber
print(f"Addition by function: {add(5, 3)}")

#subtraction function
def sub(firstnumber, secondnumber):
    return firstnumber - secondnumber
print(F"Subtraction by function: {sub(7, 8)}")

#multiplication function 
def multi(firstnumber, secondnumber):
    return firstnumber * secondnumber
print(f"Multiplication by function: {multi(5, 3)}")

#division function 
def div(firstnumber, secondnumber):
    return firstnumber / secondnumber
print(f"Division by function: {div(12,4)}")

#division function hanle zero
def div(firstnumber, secondnumber):
    if secondnumber == 0:
        return "Cannot divide by zero"
    return firstnumber / secondnumber
print(f"Division by function: {div(16,4)}")
print(f"Division by function and hanle zero: {div(12, 0)}")

#power function 
def pwr(firstnumber, secondnumber):
    return firstnumber ** secondnumber
print(f"Power by Function: {pwr(2, 3)}")

#rectangle area
def area(length,width):
    return length * width
print("Area of Rectangle:",area(5 ,5))

#rectabgle perimeter
def perimeter(lenth, width):
    return 2 * (lenth + width)
print("perimeter of rectangle by function: ",perimeter(2,4))
```

```python
#practice for arithmetic function

#1️⃣ Sum of squares of two numbers
def sum_of_squares(firstnumber, secondnumber):
    return firstnumber**2 + secondnumber**2
print(f"Sum of squares: {sum_of_squares(3, 4)}")

#2️⃣ Difference of cubes of two numbers
def difference_of_cubes(firstnumber, secondnumber):
    return firstnumber**3 - secondnumber**3
print(f"Difference of cubes: {difference_of_cubes(5, 2)}")

#3️⃣ Add three number and input we will take from user.
def add_three_Number(a, b, c):
    return a + b + c

#In def function we can not use input function directly 
firstnumber = int(input(f"Enter your value for firstnumber : "))
second_number = int(input(f"Enter your value for second number: "))
third_number = int(input(f"Enter your value for thrid number: "))

Sum_of_Three = add_three_Number(firstnumber, second_number, third_number)
print(Sum_of_Three)

#4 Area of Triangle

base = 5
height = 10

#formula for find out area of triangle
area = int(1/2 * base * height)

print("Area of Triangle: ",area)

#5 Area of Trapezium

a = 5
b = 5
h = 10

#formula for find out area of Trapezium
area = 0.5*(a + b)* h # same as (1/2) * base * height

print("Area of Trapezium: ",area)

#6 Area of cirle 

#import math function into python
import math


r = 6

#formula for area of circle 

area = math.pi * r ** 2# You can also write direct 22/7 or 3.14159 for the math.pi 

print("Area of Circle: ",area) 

#7 Convert Km into Miles or Miles into Km
# 1 KM = 0.621371 Miles
# 1 Miles = 1.60934 KM

#Formula to get Miles from Km
Km = 4

Miles = Km * 0.621371

Kilometers = 1.60934 * Miles

print("Miles: ", Miles)
print("Kilometers: ",Kilometers)

#Formula for find displacement or we can call distance 

#initial velocity(start point)
u = 5

#moving with some speed or accleration let's call as a
a = 10

#final velocity (end point)
v = 20

#formula to find distance or displacement

d = (v * v - u * u) / (2 * a)

print("Distance between two points: ",d)

```
