```python
# Reverse a number and check palindrome

n = int(input("Enter your number: "))
original = n   # store original number

reversed_num = 0

while n > 0:
    digit = n % 10
    reversed_num = reversed_num * 10 + digit
    n //= 10

print("Reversed number:", reversed_num)

if original == reversed_num:
    print("Your number is a palindrome")
else:
    print("Your number is not a palindrome")
```

```python
# Exercise 1: Countdown Timer

# Print numbers from n to 1
# If number is divisible by 3 â†’ print "Boom" instead
# At the end print "Blast Off ðŸš€"

def countdown(n):
    while n > 0:
        if n % 3 == 0:
            print("Boom")
        else:
            print(n)
        n -= 1

    print("Blast Off ðŸš€")


countdown(6)

```

```python
# Exercise 2: Login System 
# Ask user for password
# Correct password = "python123"
# Allow only 3 attempts
# If correct â†’ print "Login Successful"
# Else after 3 attempts â†’ print "Account Locked"

def attmp(n):
    while n > 0:
        password = input("Enter your Password: ")

        if password == "python123":
            print("Login Successful")
            return   # Exit function immediately
        else:
            print("Wrong Password")
            n -= 1
            print("Attempts left:", n)

    print("Account Locked")


attmp(3)





```

```python
# Exercise 3: Even Number Collector
# Keep asking user numbers
# Stop when user enters 0
# Store only EVEN numbers

def collector():
    even_numbers = []

    while True:
        num = int(input("Enter a number (0 to stop): "))

        if num == 0:
            print("Stopping input...")
            break

        if num % 2 == 0:
            even_numbers.append(num)

    print("Even numbers collected:", even_numbers)


collector()
```

```python
# Exercise 4: ATM Machine Simulation

# Start balance = 5000
#markdown
    # 1. Check Balance
    # 2. Deposit
    # 3. Withdraw
    # 4. Exit

#Rules:
    # If withdraw amount > balance â†’ print "Insufficient Funds"
    # Use if conditions properly
    # Break loop when Exit selected


Balance = 5000  # Starting balance is fixed, not user input

while True:
    print("\n--- ATM Menu ---")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == "1":
        print(f"Your current balance is: {Balance}")

    elif choice == "2":
        amount = int(input("Enter deposit amount: "))
        Balance = Balance + amount
        print(f"Successfully deposited {amount}. New balance: {Balance}")

    elif choice == "3":
        amount = int(input("Enter withdrawal amount: "))
        if amount > Balance:
            print("Insufficient Funds")
        else:
            Balance = Balance - amount
            print(f"Successfully withdrew {amount}. New balance: {Balance}")

    elif choice == "4":
        print("Thank you for using our ATM. Goodbye!")
        break

    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")
```

```python
# Exercise 11
# Level Intermediate
# Reverse a Number
# Ask the user for a number. Use a while loop to reverse its digits and print the result. (e.g. 1234 â†’ 4321)

num = int(input("Enter your number: "))
original = num

reversed_num = 0             # âœ… This builds the reversed number

while num > 0:               # âœ… Keep going while digits remain
    digit = num % 10         # Extract last digit
    reversed_num = (reversed_num * 10) + digit  # Build reversed number
    num = num // 10          # Remove last digit

print(f"Original number: {original}")
print(f"Reversed number: {reversed_num}")

    
```

```python
# Exercise 12
# Level Advance
# Simple Calculator Loop
# Build a calculator that keeps running, asking the user for two numbers and an operator (+, -, *, /). It should only stop when the user types quit.

arth = ['+', '-', '*', '/', '//', '%', '**']

while True:
    your_1st_number = input('Enter your First Number (or quit to exit): ')
    if your_1st_number == 'quit':       # âœ… Check quit BEFORE converting to int
        break

    your_2nd_number = input('Enter your Second Number (or quit to exit): ')
    if your_2nd_number == 'quit':
        break

    Operation = input(f'Which operation? {arth}: ')
    if Operation == 'quit':
        break

    num1 = int(your_1st_number)         # âœ… Convert AFTER quit check
    num2 = int(your_2nd_number)

    if Operation == '+':                # âœ… Use if/elif to perform correct operation
        result = num1 + num2
    elif Operation == '-':
        result = num1 - num2
    elif Operation == '*':
        result = num1 * num2
    elif Operation == '/':
        if num2 == 0:                   # âœ… Always check division by zero
            print('Error: Cannot divide by zero!')
            continue
        result = num1 / num2
    elif Operation == '//':
        result = num1 // num2
    elif Operation == '%':
        result = num1 % num2
    elif Operation == '**':
        result = num1 ** num2
    else:
        print(f'Invalid operation! Please choose from {arth}')
        continue

    print(f'{num1} {Operation} {num2} = {result}')

print('Goodbye! Calculator closed.')
```

```python
# Exercise 12
# Level Advance
# Prime Number Checker
# Ask the user for a number and use a while loop to check if it is prime. Keep asking for new numbers until they type stop.

while True:
    user_input = input("Enter a number (or 'stop' to exit): ")
    
    if user_input == 'stop':
        print("Goodbye!")
        break
    
    num = int(user_input)
    
    if num < 2:                        # 0 and 1 are not prime by definition
        print(f"{num} is NOT a Prime Number")
        continue
    
    i = 2
    is_prime = True                    # Assume prime until proven otherwise
    
    while i < num:
        if num % i == 0:               # Divisible by something other than 1 and itself
            is_prime = False
            break                      # No need to check further
        i += 1
    
    if is_prime:
        print(f"{num} IS a Prime Number âœ…")
    else:
        print(f"{num} is NOT a Prime Number âŒ")
```

```python
# Exercise 12
# Level Advance
# ATM Upgrade Extend
# ATM from Exercise 4 to also track a transaction history and print all past transactions when the user selects a new option 5 called "Transaction History".

Balance = 5000
transaction_history = []               # âœ… Empty list to store all transactions

while True:
    print("\n--- ATM Menu ---")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Transaction History")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        print(f"Your current balance is: {Balance}")

    elif choice == "2":
        amount = int(input("Enter deposit amount: "))
        Balance = Balance + amount
        transaction_history.append(f"Deposited:  +{amount}  | Balance: {Balance}")  # âœ… Track it
        print(f"Successfully deposited {amount}. New balance: {Balance}")

    elif choice == "3":
        amount = int(input("Enter withdrawal amount: "))
        if amount > Balance:
            print("Insufficient Funds!")
        else:
            Balance = Balance - amount
            transaction_history.append(f"Withdrew:   -{amount}  | Balance: {Balance}")  # âœ… Track it
            print(f"Successfully withdrew {amount}. New balance: {Balance}")

    elif choice == "4":
        print("\n--- Transaction History ---")
        if len(transaction_history) == 0:      # âœ… Handle empty history
            print("No transactions yet.")
        else:
            i = 1
            while i <= len(transaction_history):           # âœ… While loop to print history
                print(f"  {i}. {transaction_history[i-1]}")
                i += 1

    elif choice == "5":
        print("Thank you for using our ATM. Goodbye!")
        break

    else:
        print("Invalid choice. Please enter 1 to 5.")
```
